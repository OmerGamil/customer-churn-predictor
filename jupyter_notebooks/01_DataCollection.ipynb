## 1. Import Libraries

```python
import os
import pandas as pd

# Set working directory to project root so all relative paths resolve correctly
os.chdir('/workspace/customer-churn-predictor')  # adjust if running locally
print("Working directory:", os.getcwd())
```

## 2. Load Dataset

```python
RAW_PATH = 'outputs/datasets/collection/WA_Fn-UseC_-Telco-Customer-Churn.csv'

df = pd.read_csv(RAW_PATH)
print(f"Dataset loaded: {df.shape[0]} rows, {df.shape[1]} columns")
```

## 3. Dataset Shape and First Rows

```python
print("Shape:", df.shape)
df.head()
```

## 4. Column Data Types

```python
df.dtypes
```

## 5. Descriptive Statistics

```python
df.describe(include='all')
```

## 6. Missing Values Summary

```python
missing = df.isnull().sum()
missing_pct = (missing / len(df)) * 100
missing_summary = pd.DataFrame({'Missing Count': missing, 'Missing %': missing_pct})
print(missing_summary[missing_summary['Missing Count'] > 0])
print("\nNote: TotalCharges may contain blank strings rather than NaN values.")
print("Blank count in TotalCharges:", (df['TotalCharges'].str.strip() == '').sum())
```

## 7. Target Variable Distribution

```python
churn_counts = df['Churn'].value_counts()
churn_pct = df['Churn'].value_counts(normalize=True) * 100
print("Churn distribution:\n")
print(pd.DataFrame({'Count': churn_counts, 'Percentage': churn_pct.round(1)}))
```

## 8. Save Raw Dataset

```python
OUT_PATH = 'outputs/datasets/collection/telco_churn_raw.csv'
df.to_csv(OUT_PATH, index=False)
print(f"Raw dataset saved to: {OUT_PATH}")
```

---
## Conclusions
- Dataset contains 7,043 customer records with 21 columns (including target).
- `TotalCharges` is stored as an object dtype and contains blank strings for a small number of rows — this will be addressed in the Data Cleaning notebook.
- Approximately 26.5% of customers churned, indicating a class imbalance that will need to be handled during modelling.
- Raw dataset saved successfully as `telco_churn_raw.csv`.
