# Notebook 01 — Data Collection

## Objectives
- Download the raw Telco Customer Churn dataset from Kaggle
- Inspect its structure, column types and data quality
- Save a raw baseline copy for use in subsequent notebooks

## Inputs
- Kaggle dataset: `blastchar/telco-customer-churn`
- `kaggle.json` credentials file placed in the project root

## Outputs
- `outputs/datasets/collection/WA_Fn-UseC_-Telco-Customer-Churn.csv`
- `outputs/datasets/collection/telco_churn_raw.csv`

---

## 1. Change Working Directory

```python
import os

# Move up one level if we are inside jupyter_notebooks/
os.chdir('..') if os.path.basename(os.getcwd()) == 'jupyter_notebooks' else None
os.makedirs('outputs/datasets/collection', exist_ok=True)
print('Working directory:', os.getcwd())
```

## 2. Install Kaggle and Set Credentials

```python
import os

# Make Kaggle look for credentials in the project root
os.environ['KAGGLE_CONFIG_DIR'] = os.getcwd()

# Set correct permissions on the credentials file (required by Kaggle)
kaggle_path = os.path.join(os.getcwd(), 'kaggle.json')
os.chmod(kaggle_path, 0o600)
print('Kaggle credentials configured.')
```

## 3. Download Dataset from Kaggle

```python
import kaggle

kaggle.api.authenticate()

kaggle.api.dataset_download_files(
    'blastchar/telco-customer-churn',
    path='outputs/datasets/collection/',
    unzip=True
)

print('Dataset downloaded successfully.')
print('Files:', os.listdir('outputs/datasets/collection/'))
```

## 4. Import Libraries

```python
import pandas as pd
```

## 5. Load Dataset

```python
df = pd.read_csv('outputs/datasets/collection/WA_Fn-UseC_-Telco-Customer-Churn.csv')
print(f'Shape: {df.shape}')
df.head()
```

## 6. Data Types and Statistics

```python
df.dtypes
```

```python
df.describe(include='all')
```

## 7. Missing Values

```python
# Standard null check
print(df.isnull().sum())

# TotalCharges is object dtype — check for hidden blank strings
print('Blank strings in TotalCharges:', (df['TotalCharges'].str.strip() == '').sum())
```

## 8. Target Distribution

```python
print(df['Churn'].value_counts())
print(df['Churn'].value_counts(normalize=True).mul(100).round(1))
```

## 9. Save Raw Dataset

```python
df.to_csv('outputs/datasets/collection/telco_churn_raw.csv', index=False)
print('Saved: telco_churn_raw.csv')
```
