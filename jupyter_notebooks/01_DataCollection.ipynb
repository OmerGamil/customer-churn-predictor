## 3 — Load the Dataset

The dataset is the IBM Telco Customer Churn dataset. It covers 7,043 customers and includes demographic information, subscribed services, account details, and whether the customer churned in the last month.

```python
RAW_PATH = 'outputs/datasets/collection/WA_Fn-UseC_-Telco-Customer-Churn.csv'

df = pd.read_csv(RAW_PATH)
print(f'Dataset loaded successfully: {df.shape[0]} rows and {df.shape[1]} columns')
```

## 4 — Inspect Shape and First Rows

```python
print('Shape:', df.shape)
df.head()
```

## 5 — Column Data Types

I want to check what types each column has been assigned. This will flag any columns that look numeric but are stored as objects.

```python
df.dtypes
```

## 6 — Descriptive Statistics

```python
df.describe(include='all')
```

## 7 — Missing Values

A standard isnull() check will not catch blank strings, so I am checking TotalCharges separately since it is stored as an object and may contain whitespace instead of NaN.

```python
missing = df.isnull().sum()
missing_pct = (missing / len(df)) * 100
print(pd.DataFrame({'Missing Count': missing, 'Missing %': missing_pct.round(2)}))
print('\nBlank strings in TotalCharges:', (df['TotalCharges'].str.strip() == '').sum())
```

## 8 — Target Variable Distribution

It is important to check the balance of the Churn column early. A large imbalance will need to be addressed during modelling.

```python
churn_counts = df['Churn'].value_counts()
churn_pct = df['Churn'].value_counts(normalize=True) * 100
print(pd.DataFrame({'Count': churn_counts, 'Percentage': churn_pct.round(1)}))
```

## 9 — Save Raw Dataset

```python
OUT_PATH = 'outputs/datasets/collection/telco_churn_raw.csv'
df.to_csv(OUT_PATH, index=False)
print(f'Saved to: {OUT_PATH}')
```
